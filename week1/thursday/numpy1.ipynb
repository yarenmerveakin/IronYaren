```python
#### 6. Find indices of non-zero elements from [1,2,0,0,4,0] (★☆☆)
`hint: np.nonzero`
```

```python
lst = [1, 2, 0, 0, 4, 0]
nonzero_indices = []

for i in range(len(lst)):
    if lst[i] != 0:
        nonzero_indices.append(i)

print(nonzero_indices)
```
