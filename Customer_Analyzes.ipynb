```python
def clean_gender(x):
    if x in ['M', "Male"]:
        return 'Male'
    elif x in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
file5['GENDER'] = list(map(clean_gender, file5['GENDER'])) 
```

```python
def clean_ST(x):
    if x in ['WA', "Washington"]:
        return 'Washington'
    elif x in ['Arizona', 'AZ']:
        return 'Arizona'
    elif x in ['California', 'Cali']:
        return 'California'
    elif x in ['Oregon']:
        return 'Oregon'
    elif x in ['Nevada']:
        return 'Nevada'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
file5['ST'] = list(map(clean_ST, file5['ST'])) 
```
