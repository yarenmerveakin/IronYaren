```python
def clean_gender(x):
    if x in ['M', "Male"]:
        return 'Male'
    elif x in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
file5['gender'] = list(map(clean_gender, file5['gender'])) 
```

```python
def clean_st(x):
    if x in ['WA', "Washington"]:
        return 'Washington'
    elif x in ['Arizona', 'AZ']:
        return 'Arizona'
    elif x in ['California', 'Cali']:
        return 'California'
    elif x in ['Oregon']:
        return 'Oregon'
    elif x in ['Nevada']:
        return 'Nevada'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
file5['st'] = list(map(clean_ST, file5['st'])) 
```
