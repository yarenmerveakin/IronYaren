```python
def clean_vehicle_class(x):
    if x in ['Luxury SUV']:
        return 'Luxury Vehicle'
    elif x in ['Luxury Car']:
        return 'Luxury Vehicle'
    
```

```python
def IQR_outliers(file5):

     Q1=file5.quantile(0.25)
     Q3=file5.quantile(0.75)
     IQR=Q3-Q1
     file5=file5[~((file5<(Q1-1.5*IQR)) | (file5>(Q3+1.5*IQR)))]
     return file5
    
IQR_outliers(file5['customer lifetime value'])
IQR_outliers(file5['income'])
IQR_outliers(file5['monthly premium auto'])
IQR_outliers(file5['number of open complaints'])
IQR_outliers(file5['total claim amount'])
```
